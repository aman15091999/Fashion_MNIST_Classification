```python
# Libraries Required
import tensorflow.keras as tk
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Data
(X_train, y_train), (X_test, y_test) = tk.datasets.fashion_mnist.load_data()
```

```python
print(X_train.shape , X_test.shape, y_train.shape, y_test.shape)
```

```python
X_train[0]
```

```python
y_train[0]
```

```python
# to show image from array using matplotlib.
plt.imshow(X_train[0], cmap='Greys');
```

```python
class_labels = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
# To get 10 image.
plt.figure(figsize=(16,16))
for i in range(10):
  plt.subplot(2,5,i+1)
  plt.imshow(X_train[i], cmap='Greys')
  # plt.axis('off')
  plt.title(class_labels[y_train[i]] + '=' + str(y_train[i]), fontsize=10)
```

```python
# Scale data(Feaute Scaling)
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[0]
```

```python
#Build Neural Network
model = tk.models.Sequential([
                      tk.layers.Flatten(input_shape=[28,28]),
                      tk.layers.Dense(units=32, activation='relu'),
                      tk.layers.Dense(units=10, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=1) #epochs apne hi result dataset ko again-and-again(iterate) fit krta rahega , jisse accuracy increase hoti rahegi.
```

```python
model.fit(X_train, y_train, epochs=20)
```

```python
# Test and Evaluate Neural Network Model
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
X_test.shape
```

```python
y_pred[0]
```

```python
y_pred[0].round(2)
```

```python
pred_idx = np.argmax(y_pred[0])
```

```python
class_labels[pred_idx] # predicted Value
```

```python
class_labels[y_test[0]] # Actual Value
```

```python
plt.figure(figsize=(16,16))
for i in range(10):
  plt.subplot(2,5,i+1)
  plt.imshow(X_test[i], cmap='Greys')
  # plt.axis('off')
  plt.title('Actual = {}\nPredicted = {}'.format(class_labels[y_test[i]], class_labels[np.argmax(y_pred[i])]))
```

```python
# Evaluation
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, [np.argmax(i) for i in y_pred])
```

```python
# Classification Report
from sklearn.metrics import classification_report
```

```python
class_report = classification_report(y_test, [np.argmax(i) for i in y_pred], target_names = class_labels)
```

```python
print(class_report)
```
